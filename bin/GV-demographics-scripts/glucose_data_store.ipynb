```python
import shutil, os
from pathlib import Path
import os.path

path = r'' # Specify directory path
dest = r'' # Specify destination path

def find_files(filename, search_path):
    result = []

# Wlaking top-down from the root
    for root, dir, files in os.walk(search_path):
        if filename in files:
            result.append(os.path.join(root, filename))
    return result

os.chdir(path)

for p,n,f in os.walk(os.getcwd()):
    for a in f:
        a = str(a)
        folder = list(filter(lambda b: 'entries' in b, n))
        if a.__contains__("_entries_"): 
            if a.endswith(".json"):
                file_source = find_files("%s.csv"%a, path)
                destination = Path(r"%s\%s.csv"%(dest, a))
                print(a)
                shutil.copy2(file_source[0], destination, follow_symlinks=False)
```
